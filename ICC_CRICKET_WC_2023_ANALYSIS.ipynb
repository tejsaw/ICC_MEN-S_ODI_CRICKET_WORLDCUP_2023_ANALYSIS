# **Which team has won most of the matches while chasing and while defending?**

```python
def innings_team(x):
    if x == "wickets":
        return "chasing"
    elif x == "runs":
        return "defending"
```

```python
df['innings_team'] = np.vectorize(innings_team)(df['won_by'])
df
```

```python
df.groupby(['winner','innings_team']).agg({"innings_team":"count"})
```

```python
px.histogram(df, x = "winner", color ='innings_team', barmode = "group")
```

# **Conclusion:**

1) Out of 10 win, India has won 5 Matches while Chasing and 5 Matches while Defending.

2) Australia has won 5 Matches while chasing and 4 while defending.

3) South Africa on the other has won 5 Matches while Defending.

Hence, it can be concluded from the above observation that India was one of the best balanced team who can chase and defend their target.

# **Are there any teams who has won both the toss and the match?**

```python
def toss_won_match_won(a,b,c,d):
    if a==c==d:
        return a
    elif b==c==d:
        return b
    else:
        return 'other'
```

```python
df['toss_won_match_won']= np.vectorize(toss_won_match_won)(df['team1'],df['team2'],df['toss_winner'],df['winner'])
```

# **How many batsman and bowler won the player of the match award?**

```python
def player_of_the_match_type(a,b,c):
    if a == b:
        return "batsman"
    elif a == c:
        return "bowler"
    elif a != b or a!= c:
        return a
```

```python
df['player_of_the_match_type'] = np.vectorize(player_of_the_match_type)(df['player_of_the_match'],df['top_scorer'],df['best_bowler'])
```

```python
df['player_of_the_match_type'].value_counts()
```

```python
# It can be noted that Travis Head, Rachin Ravindra and Aiden Markram are batsman and Mujeeb Ur Rahman is a bowler, Hence replacing their designation in place of their Name.
df['player_of_the_match_type'] = df['player_of_the_match_type'].replace(['Travis Head','Rachin Ravindra', 'Aiden Markram','Mujeeb Ur Rahman'],['batsman','batsman','batsman','bowler'])
```

```python
df['player_of_the_match_type'].value_counts()
```

```python
px.pie(df, names = 'player_of_the_match_type')
```

# ***Conclusion:***

In 48 Matches, 31 times the batsman won the player of the match award and 17 times the winner of the player of the match award was a bowler.

Hence, it is safe to say that the batsman dominated in the tournament as compared to the bowler
